```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
%matplotlib inline
```

```python
data = pd.read_csv("data/monsters.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data = data[data.legendary_actions.notna() == False]
# data = data[data.condition_immunities == '[]']
#data = data[data.legendary_actions == None]
```

```python
data.head()
```

```python
data.describe()
```

```python
train1 = data[['armor_class','hit_points','strength','dexterity','constitution','intelligence','wisdom']]
```

```python
train1.head()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
labels = data['challenge_rating']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
```

```python
y_pred = reg.predict(train1)
```

```python
print('predicted response:', y_pred, sep='\n')
```
